### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
- 「File I/O」只是把 XML 內容呈個整現出來 (dump), 無法針對特定某一筆資料 (節點內容及屬性) 去做存取
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
import xmltodict

# 1. 請問高雄市有多少地區有溫度資料？

with open("../data/64_72hr_CH.xml", encoding='utf-8-sig') as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
print(len(doc['cwbopendata']['dataset']['locations']['location']))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for node in doc['cwbopendata']['dataset']['locations']['location']:
    area = node['locationName']
    time = node['weatherElement'][0]['time'][0]['dataTime']
    tempNode = node['weatherElement'][0]['time'][0]['elementValue']
    temp = tempNode['value'] + ' ' + tempNode['measures']
    print('地點 : {:<8} ==> 時間 : {:<25} / 溫度: {:<10}'.format(area, time, temp))
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
firstAreaNode = doc['cwbopendata']['dataset']['locations']['location'][0]
print(firstAreaNode['locationName'] + '-' + firstAreaNode['weatherElement'][0]['description'])
for node in firstAreaNode['weatherElement'][0]['time']:
    time = node['dataTime']
    tempNode = node['elementValue']
    temp = tempNode['value'] + ' ' + tempNode['measures']
    print('時間 : {:<25} / 溫度: {:<10}'.format(time, temp))
```
