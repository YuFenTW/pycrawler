## requests + BeautifulSoup


> [Python 使用 Beautiful Soup 抓取與解析網頁資料，開發網路爬蟲教學](https://blog.gtwang.org/programming/python-beautiful-soup-module-scrape-web-pages-tutorial/)

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'

html_doc = r.text
soup = BeautifulSoup(html_doc, "html5lib")

# 【方法 1】取得 meta title
# title = soup.find("meta",  property="og:title")
# print(type(title), title['content'])

# 【方法 2】取得 meta title 
title = soup.title
print(type(title), title.string)
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq

g = Grab(timeout=10, user_agent='grab')

resp = g.go('https://www.dcard.tw/f')
resp.body
```

```python
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
```
